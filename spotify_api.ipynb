```python
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json

# Spotify API credentials
client_id = '9a2d1bf8571f4f8ab5e5d7d00f25bda4'
client_secret = 'e87f635008934938b7482140ba158674'

# Spotify API endpoint to get access token
url = 'https://accounts.spotify.com/api/token'

# sending a POST request to spotify API with client credentials to obtain access token
auth_response = requests.post(url, {
    'grant_type': 'client_credentials',
    'client_id': client_id,
    'client_secret': client_secret,
})


# convert the response to JSON format
auth_response_data = auth_response.json()

# extract the access token from the response
access_token = auth_response_data['access_token']

# create headers for authorization using the obtained access token
headers = {'Authorization': 'Bearer {token}'.format(token=access_token)}

# base URL for Spotify API endpoints
base_url = 'https://api.spotify.com/v1/'

# Spotify artist id for a specific artist
artist_id = '246dkjvS1zLTtiykXe5h60'

# pull all artists albums
r = requests.get(base_url + 'artists/' + artist_id + '/albums', 
                 headers=headers, 
                 params={'include_groups': 'album', 'limit': 50})

# converting the response into JSON format
d = r.json()


# iterate through each album in the response and print its name and release date
for album in d['items']:
    print(album['name'], ' --- ', album['release_date'])
```

```python
# Create an empty DataFrame to store the song information
song_df = pd.DataFrame(columns=['Album Name', 'Release Date', 'Song Name', 'Song URI'])

# Iterate through the albums
for album in d['items']:
    album_name = album['name']
    release_date = album['release_date']

    # Get the songs in each album
    album_id = album['id']
    r = requests.get(base_url + 'albums/' + album_id + '/tracks',
                     headers=headers,
                     params={'limit': 50})
    tracks = r.json()['items']

    # Iterate through the songs in the album
    for track in tracks:
        song_name = track['name']
        song_uri = track['uri']
        
        # Additional features request
        features_url = 'https://api.spotify.com/v1/audio-features/' + track['id']
        features_response = requests.get(features_url, headers=headers)
        features_data = features_response.json()
        
         # Extract additional features
        acousticness = features_data['acousticness']
        danceability = features_data['danceability']
        energy = features_data['energy']
        instrumentalness = features_data['instrumentalness']
        liveness = features_data['liveness']
        loudness = features_data['loudness']
        speechiness = features_data['speechiness']
        tempo = features_data['tempo']
        valence = features_data['valence']
        
        # Append the song information to the DataFrame
        song_df = song_df.append({'Album Name': album_name,
                                  'Release Date': release_date,
                                  'Song Name': song_name,
                                  'Song URI': song_uri,
                                  'Acousticness' : acousticness,
                                  'Danceability' : danceability,
                                  'Energy': energy,
                                  'Instrumentalness' : instrumentalness,
                                  'Tempo' : tempo,
                                  'Valence' : valence},
                                  ignore_index=True)

df = pd.DataFrame(song_df)

df
```
